# **Média de notas + aprovado/reprovado**

```python
notas = [8, 7, 6, 9]
soma = 0

for n in notas:
    soma += n

media = soma / len(notas)

print("Média:", media)

if media >= 7:
    print("Aprovado!")
else:
    print("Reprovado!")
```

# **Função que soma uma lista**

```python
def soma_lista(lista):
    soma = 0
    for num in lista:
        soma += num
    return soma

print(soma_lista([1,2,3,4]))
```

# **Função que verifica palíndromo**

```python
def palindromo(palavra):
    palavra = palavra.lower()
    return palavra == palavra[::-1]

print(palindromo("arara"))
print(palindromo("casa"))
```

# **Estatísticas no NumPy**

```python
import numpy as np

# Criando um array de exemplo
dados = np.array([10, 20, 30, 40, 50])

print("Dados:", dados)

# --- Estatísticas básicas ---
media = np.mean(dados)
mediana = np.median(dados)
desvio = np.std(dados)
variancia = np.var(dados)
minimo = np.min(dados)
maximo = np.max(dados)

print("\n--- Estatísticas ---")
print("Média:", media)
print("Mediana:", mediana)
print("Desvio Padrão:", desvio)
print("Variância:", variancia)
print("Valor mínimo:", minimo)
print("Valor máximo:", maximo)

# --- Estatísticas por linha e coluna ---
matriz = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("\nMatriz:")
print(matriz)

print("\nMédia por coluna (axis=0):", np.mean(matriz, axis=0))
print("Média por linha  (axis=1):", np.mean(matriz, axis=1))

print("\nSoma por coluna (axis=0):", matriz.sum(axis=0))
print("Soma por linha  (axis=1):", matriz.sum(axis=1))
```
